# <a id='toc2_'></a>[Gráfico de líneas básico con lineplot](#toc0_)

La función **lineplot** de seaborn permite crear gráficos de líneas en Python. Tan solo necesitas pasar tus datos a la función para crear un gráfico básico con una línea azul sólida por defecto.

```python
import numpy as np

# Semilla
rng = np.random.RandomState(0)

# Simulación de datos
x = np.array(range(0, 20))
y = np.square(x) + rng.uniform(0, 100, 20)

# Conjunto de datos
df = {'x': x, 'y': y}
```

```python
import seaborn as sns

sns.lineplot(x=x,y=y)
```

```python
# Equivalente a:
sns.lineplot(x = "x", y = "y", data = df)
```

## <a id='toc2_1_'></a>[Color](#toc0_)

El argumento color controla el color de la línea del gráfico. Puedes seleccionar el color que quieras para tu visualización.

```python
import seaborn as sns

sns.lineplot(x = "x", y = "y", data = df,
             color = "red")
```

## <a id='toc2_2_'></a>[Línea discontinua y ancho de la línea](#toc0_)

El estilo de la línea se puede establecer con el argumento linestyle. En el siguiente ejemplo estamos creando un gráfico con una línea discontinua, pero también se puede crear una línea punteada pasando "dotted".
Por último, si se quiere modificar el grosor por defecto de la línea se puede pasar cualquier valor positivo a linewidth, tal y como se muestra a continuación.

```python
import seaborn as sns

sns.lineplot(x = "x", y = "y", data = df,
            linestyle = "dashed", linewidth=3)
```

# <a id='toc3_'></a>[Gráfico de barras](#toc0_)
La función **countplot** permite representar el número de observaciones para cada grupo de una variable categórica mediante barras. Por eso solo pasamos el valor de **x**.

```python
import seaborn as sns

data = ["A", "A", "B",
        "B", "B", "C"]

# Count plot
sns.countplot(x = data) # Cuenta las ocurrencias de cada valor: 2 para A, 3 para B y 1 para C
```

### <a id='toc3_1_1_'></a>[Count plot horizontal](#toc0_)

En caso de que se quiera crear un count plot horizontal se debe pasar los datos al argumento y, en lugar de a x.

```python
import seaborn as sns

data = ["A", "A", "B",
        "B", "B", "C"]

# Count plot
sns.countplot(y = data)
```

### <a id='toc3_1_2_'></a>[Grupos y subgrupos](#toc0_)

Además, si se tiene otra variable categórica que representa subgrupos es posible pasarla al argumento hue, tal y como se muestra en el siguiente ejemplo.    
Establecemos dodge = False en caso de que se quiera que las barras se solapen para cada grupo.

```python
import seaborn as sns

data = ["A", "A", "B", "A",
        "B", "B", "C", "C"]
group = ["G1", "G1", "G2", "G2",
         "G1", "G2", "G1", "G2"]

sns.countplot(x = data, hue = group) # Cuenta las ocurrencias de cada valor: 3 para A, 3 para B y 2 para C y los agrupa por G1 y G2
```

### <a id='toc3_1_3_'></a>[Paleta de colores](#toc0_)

Si se quiere sobrescribir la paleta de colores por defecto, se puede cambiar con el argumento **palette**, tal y como se muestra en el ejemplo siguiente.

```python
import seaborn as sns

data = ["A", "A", "B",
        "B", "B", "C"]

# sns.countplot(x = data, palette = "Set1") # deprecated!!
sns.countplot(x = data, hue=data, legend=False, palette="Set1")
```

### <a id='toc3_1_4_'></a>[Un único color para todas las barras](#toc0_)

También es posible establecer el mismo color para todas las barras pasando el nombre de un color al argumento color de la función.

```python
import seaborn as sns

data = ["A", "A", "B",
        "B", "B", "C"]

# Count plot
sns.countplot(x = data,
              color = "lightblue")
```

### <a id='toc3_1_5_'></a>[Color del borde](#toc0_)

Por último, también es posible modificar los colores de los bordes de cada barra, estableciendo un color o un vector de colores con el argumento edgecolor.

```python
import seaborn as sns

data = ["A", "A", "B",
        "B", "B", "C"]

# Count plot
sns.countplot(x = data,
              edgecolor = "red")
```

## <a id='toc3_2_'></a>[Otra opción para gráficos de barras](#toc0_)

La función **barplot** permite crear gráficos de barras en seaborn. Esta función *calcula la media* para cada variable categórica y la representa con barras, así como sus intervalos de confianza con barras de error.     

```python
import seaborn as sns
tips = sns.load_dataset('tips') # Importar el dataset tips de seaborn

sns.barplot(x='sex',y='total_bill',data=tips) # Barplot de la media de la variable "total_bill" agrupada por la variable "sex"
```

```python
import numpy as np
import seaborn as sns
import random

# Simulación de datos
rng = np.random.RandomState(0)
variable = rng.normal(20, 1, size = 50) # Se generan 50 valores aleatorios con media 20 y desviación estándar 1
random.seed(0)
grupo = random.choices(["G1", "G2", "G3"], k = 50) # Se generan 50 valores aleatorios de los grupos G1, G2 y G3
grupo2 = random.choices(["A", "B"], k = 50)        # Se generan 50 valores aleatorios de los grupos A y B
df = {'variable': variable, 'grupo': grupo, 'grupo2': grupo2}  # Se crea un diccionario con las variables, a modo de dataset
```

```python
# Gráfico de barras
sns.barplot(x = grupo, y = variable)
```

```python
# Equivalente a:
sns.barplot(x = "grupo", y = "variable", data = df)
```

## <a id='toc3_3_'></a>[Orden](#toc0_)

Se debe tener en cuenta que el orden por defecto de las barras depende del orden de aparición de los grupos de la variable. Si se quiere sobrescribir el orden se debe pasar una lista con el orden que se quiera al argumento order.

```python
import seaborn as sns

sns.barplot(x = grupo, y = variable,
            order = ["G1", "G2", "G3"])
```


## <a id='toc3_4_'></a>[Eliminar las barras de error](#toc0_)

Las barras de error se muestran sobre cada barra por defecto, pero es posible eliminarlas estableciendo el argumento ci como False. También se puede establecer este argumento como "sd" de forma que las barras de error se calcularán en base a la desviación típica de los datos.

```python
import seaborn as sns

sns.barplot(x = grupo, y = variable,
            ci = None)
```

## <a id='toc3_5_'></a>[Agregar segmentos horizontales a las barras de error](#toc0_)

Si se quiere añadir los segmentos horizontales que acotan las barras de error se puede seleccionar su tamaño con el argumento capsize, tal y como se muestra a continuación.

```python
import seaborn as sns

sns.barplot(x = grupo, y = variable,
            capsize = 0.25)
```

## <a id='toc3_6_'></a>[Paleta de colores](#toc0_)

Una alternativa es seleccionar una paleta de colores o pasar un diccionario con un color para cada grupo al argumento palette de la función.

```python
import seaborn as sns

sns.barplot(x = grupo, y = variable,
            palette = "Blues_d")
```

## <a id='toc3_7_'></a>[Personalización avanzada](#toc0_)

Por último, cabe destacar que existen otros argumentos para personalizar el estilo del gráfico, como edgecolor, para cambiar el color de los bordes, entre otros.



```python
import seaborn as sns

sns.barplot(x = grupo, y = variable,
            linewidth = 1,
            facecolor = "white",
            edgecolor = "red")
```

# <a id='toc4_'></a>[Gráfico de dispersión](#toc0_)

Para crear un gráfico de dispersión en Python con seaborn se puede usar la función scatterplot o relplot. Existen varias opciones para pasar los datos: se puede pasar las variables para los ejes X e Y a los argumentos x e y como arrays, pasar un conjunto de datos al argumento data y establecer los nombres de las variables o pasar un data frame al argumento data con el formato adecuado. Los símbolos serán azules por defecto.

Datos de muestra

```python
import numpy as np

# Semilla
rng = np.random.RandomState(0)

# Simulación de datos
x = rng.uniform(0, 1, 500)
y = 5 * x + rng.normal(0, 2, size = 500)
group = np.where(x < 0.4, "A", np.where(x > 0.8, "C", "B"))
x = x + rng.uniform(-0.2, 0.2, 500)

# Conjunto de datos
df = {'x': x, 'y': y, 'group': group}
```

```python
import seaborn as sns

sns.scatterplot(x = x, y = y)

# Equivalente a:
sns.scatterplot(x = "x", y = "y", data = df)

# Equivalente a:
sns.relplot(x = x, y = y)
```

## <a id='toc4_1_'></a>[Color de los símbolos](#toc0_)

En caso de que se quiera sobrescribir el color por defecto de los símbolos es posible especificar un nuevo color con el argumento color de la función.

```python
import seaborn as sns

sns.scatterplot(x = x, y = y,
                color = "green")
```

## <a id='toc4_2_'></a>[Color del borde](#toc0_)

Ten en cuenta que también puedes modificar el color de los símbolos con edgecolor, que por defecto es blanco. En caso de que quieras eliminar los bordes establece edgecolor = "none", edgecolor = None o linewidth = 0.

```python
import seaborn as sns

sns.scatterplot(x = x, y = y,
                edgecolor = "red")
```

## <a id='toc4_3_'></a>[Transparencia](#toc0_)

El argumento alpha de la función controla la transparencia de los símbolos. Esto es muy útil cuando hay muchos datos y los puntos se superponen unos encima de otros. De esta manera, se pueden ver las zonas de mayor densidad.   

(Utilidad similar al gráfico de densidad hexagonal)

```python
import seaborn as sns

sns.scatterplot(x = x, y = y,
                hue = y)
```

## <a id='toc4_4_'></a>[Tamaño de los símbolos](#toc0_)

Dependiendo de tus datos tendrás que modificar el tamaño de los símbolos del gráfico. Por ejemplo, si tienes muchas observaciones podrías considerar reducir el tamaño de los símbolos con el argumento size.

```python
import seaborn as sns

sns.scatterplot(x = x, y = y,
                size = 1, legend = False)
```

## <a id='toc4_5_'></a>[Forma de los símbolos](#toc0_)

Por defecto, los símbolos del gráfico de dispersión son círculos, pero puedes seleccionar entre una gran variedad de formas con el argumento marker. Por ejemplo, si pasamos "s" los símbolos serán cuadrados.

```python
import seaborn as sns

sns.scatterplot(x = x, y = y,
                marker = "s")
```

## <a id='toc4_6_'></a>[Stripplot - Gráfico de dispersión](#toc0_)

Este tipo de gráfico dibuja un diagrama de dispersión donde una variable es categórica.    
Un *stripplot* se puede dibujar como gráfico individual, pero es un buen complemento en los casos en que desea mostrar todas las observaciones junto con alguna representación de la distribución subyacente.

```python
sns.stripplot(x="day", y="total_bill", data=tips)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True) # el parámetro "jitter" se usa para dispersar los puntos
```

```python
# Varias variables
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1')
```

```python
# Varias Variables
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1',dodge=True); # el parámetro "dodge" se usa para separar las categorías
```

## <a id='toc4_7_'></a>[Swarmplot - Gráfico de dispersión](#toc0_)

Este gráfico es similar a **stripplot**()**, pero los puntos se ajustan (solo a lo largo del eje categórico) para que no se superpongan.   
Esto proporciona una mejor representación de la distribución de los valores, aunque no se ajusta a un gran número de observaciones (tanto en términos de la capacidad de mostrar todos los puntos como en términos del cálculo necesario para organizarlos).

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
```

```python
sns.swarmplot(x="day", y="total_bill",hue='sex',data=tips, palette="Set1", dodge=True) # el parámetro "dodge" se usa para separar las categorías
```

## <a id='toc4_8_'></a>[Gráfico de dispersión + recta regresión lineal (**lmplot**)](#toc0_)

```python
#scatter plot mas la regresion lineal
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
#scatterplot más la regresion lineal separada por sexo
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex');
```

```python
# Cambiando de paleta de colores
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm');
```

# <a id='toc5_'></a>[Gráfico de violín](#toc0_)

Un **gráfico de violín** realiza un papel similar a un diagrama de cajas y bigotes. También nos muestra la distribución de datos cuantitativos a través de varios niveles de una (o más) variables categóricas, de modo que esas distribuciones se puedan comparar.    
A diferencia de un diagrama de caja, en el que todos los componentes de la gráfica corresponden a los puntos de datos reales, la gráfica del violín *presenta una estimación de la densidad* del núcleo de la distribución subyacente.

```python
# sns.violinplot(x="day", y="total_bill", data=tips,palette='rainbow')  # deprecated!!
sns.violinplot(x="day", y="total_bill", data=tips, hue= "day", palette='rainbow')
```

```python
# Varias Variables
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',palette='Set1');
```

```python
# Varias variables
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',split=True,palette='Set1') # Se divide el violín en dos partes, una para cada valor, usando el parámetro "split"
```

# <a id='toc6_'></a>[Combinando gráficos categóricos](#toc0_)

```python
sns.violinplot(x="tip", y="day", data=tips, hue="day", palette='rainbow') # grafico de violin
sns.swarmplot(x="tip", y="day", data=tips,color='red',size=3) # grafico swarmplot
```

# <a id='toc7_'></a>[Histogramas](#toc0_)

Dado un vector, se puede crear un histograma en seaborn con la función **histplot**. Se debe tener en cuenta que se puede pasar una única variable o una variable de un conjunto de datos, tal y como se muestra en el ejemplo siguiente.

```python
import numpy as np
import seaborn as sns

# Simulación de datos
rng = np.random.RandomState(0)
x = rng.normal(0, 1, size = 1000)
df = {'x': x}

# Histograma
sns.histplot(x = x)

# Equivalente a:
sns.histplot(x = "x", data = df)
```

## <a id='toc7_1_'></a>[Número y grosor de barras o clases](#toc0_)

Cuando se crea un histograma con seaborn el número de clases (barras) se calcula automáticamente con la función numpy.histogram_bin_edges, cuyo estimador por defecto se llama "auto" y es el máximo de los estimadores "sturges" (método de Sturges) y "fd" (Freedman Diaconis). Sin embargo, existen diferentes maneras de calcular o indicar el número de barras del histograma resultante.

### <a id='toc7_1_1_'></a>[Binwidth](#toc0_)

Es posible seleccionar el ancho de las barras con el argumento binwidth, tal y como se muestra a continuación.


```python
import numpy as np
import seaborn as sns

# Simulación de datos
rng = np.random.RandomState(0)
x = rng.normal(0, 1, size = 1000)
df = {'x': x}

sns.histplot(x = x, binwidth = 0.5)
```

## <a id='toc7_2_'></a>[Color de fondo](#toc0_)

Por defecto, el color de cada barra es azul. Sin embargo, puedes elegir otro color con el argumento color. Ten en cuenta que también puedes modificar la transparencia del color con el argumento alpha de la función.

```python
import numpy as np
import seaborn as sns

# Simulación de datos
rng = np.random.RandomState(0)
x = rng.normal(0, 1, size = 1000)
df = {'x': x}

sns.histplot(x = x,
             color = "red", alpha = 0.5)
```

```python
import numpy as np
import seaborn as sns

# Simulación de datos
rng = np.random.RandomState(0)
x = rng.normal(0, 1, size = 1000)
df = {'x': x}

sns.histplot(x = x,
             fill = False)
```

# <a id='toc8_'></a>[Gráfico de cajas y bigotes - Boxplot](#toc0_)

```python
import numpy as np
import seaborn as sns
import random

# Simulación de datos
rng = np.random.RandomState(5)
variable = rng.normal(0, 1, size = 100)
random.seed(5)
grupo = random.choices(["G1", "G2", "G3"], k = 100)
grupo2 = random.choices(["A", "B"], k = 100)
df = {'variable': variable, 'grupo': grupo, 'grupo2': grupo2}
```

## <a id='toc8_1_'></a>[Gráfico de cajas](#toc0_)
Pasando una variable numérica al argumento x de la función boxplot se puede crear un gráfico de cajas y bigotes en seaborn. Es necesario tener en cuenta que los datos pueden ser una variable o una columna de un data frame.

```python
import seaborn as sns

# Box plot
sns.boxplot(x = variable)

# Equivalente a:
sns.boxplot(x = "variable", data = df)
```

## <a id='toc8_2_'></a>[Vertical](#toc0_)

Si prefieres un gráfico de cajas vertical tendrás que pasar tus datos al argumento y en lugar de a x.

```python
import seaborn as sns

# Box plot
sns.boxplot(y = variable)
```

## <a id='toc8_3_'></a>[Personalización](#toc0_)

```python
import seaborn as sns

# Box plot
sns.boxplot(x = variable,
            color = "orange",
            medianprops = {"color": "blue"},
            whiskerprops = {"color": "green"},
            capprops = {"color": "red"},
            fliersize = 10, # Tamaño de los atípicos/outliers
            linewidth = 3,  # Grosor de las líneas
            saturation = 0.95) # Saturación del color
```


## <a id='toc8_4_'></a>[Box plot por grupo](#toc0_)

Si se pasa una variable categórica y una numérica a la función se puede crear un gráfico de cajas por grupo.    
En caso de que fuera necesario crear un box plot por grupo horizontal, se debe pasar la variable numérica a **x** y la categórica a **y**.

```python
import seaborn as sns

# Box plot por grupo
sns.boxplot(x = variable,
            y = grupo)
```

```python
import seaborn as sns

# Box plot por grupo
sns.boxplot(y = variable,
            x = grupo,
            palette="Set2",
            hue = grupo2, # Agrupamos usando una variable más (grupo2)
            order=["G1", "G2", "G3"]) # Establecemos el orden
```

# <a id='toc9_'></a>[Mapa de calor](#toc0_)
La función heatmap permite crear mapas de calor o gráficos de correlación en Python con seaborn. Se puede pasar un conjunto de datos 2D o un data frame de pandas.    
En caso de utilizar un data frame de pandas, los nombres de las filas y las columnas serán usados como etiquetas para los ejes.

```python
import numpy as np
import seaborn as sns

# Simulación de datos
np.random.seed(1)
data = np.random.rand(10, 10)

sns.heatmap(data)
```

## <a id='toc9_1_'></a>[Celdas cuadradas](#toc0_)

Se debe tener en cuenta que dependiendo del número de filas y columnas y del tamaño del gráfico las celdas del mapa de calor podrían no ser cuadradas.   
Si se quiere que las celdas sean cuadradas establece el argumento square como True.

```python
import numpy as np
import seaborn as sns

# Simulación de datos
np.random.seed(1)
data = np.random.rand(10, 10)

sns.heatmap(data, square = True)
```

```python
import numpy as np
import seaborn as sns

# Simulación de datos
np.random.seed(1)
data = np.random.rand(10, 10)

sns.heatmap(data,
            linewidths = 0.75,
            linecolor = "white") # Color por defecto

```

# <a id='toc10_'></a>[Gráficos por pares ("pair plot")](#toc0_)
La función **pairplot** de seaborn permite crear gráficos por pares en Python.    
Tan solo es necesario pasar un conjunto de datos en long-format, donde cada columna es una variable y cada fila un individuo.

```python
import seaborn as sns

df = sns.load_dataset("iris")
```

```python
import seaborn as sns

sns.pairplot(df)
```

## <a id='toc10_1_'></a>[Selección de variables](#toc0_)

Es necesario tener en cuenta que también se puede seleccionar las variables que se quiera incluir en la representación con vars. En caso de querer incluir distintas variables para los ejes X e Y se puede utilizar los argumentos x_vars e y_bars en su lugar.

```python
import seaborn as sns

sns.pairplot(df, vars = ["petal_length", "petal_width"])
```

## <a id='toc10_2_'></a>[Color por grupo](#toc0_)

Si el conjunto de datos con el que se trabaja contiene una variable categórica que representa grupos es posible pasarla al argumento *hue*, de modo que las representaciones gráficas se crearán y colorearán en base a esos grupos.

```python
import seaborn as sns

sns.pairplot(df, hue = "species")
```

## <a id='toc10_3_'></a>[Personalización](#toc0_)

Los argumentos *plot_kws* y *diag_kws* se pueden utilizar para realizar una personalización más avanzada del gráfico pasando diccionarios.     
Mientras que el primero se usa para modificar los estilos de los gráficos de los paneles inferior y superior, el segundo permite personalizar el estilo de la diagonal.

```python
import seaborn as sns

sns.pairplot(df,
             plot_kws = {'color': 'green', 'marker': 's'}, # la 's' (square) es para generar puntos cuadrados
             diag_kws = {'color': 'orange'}) 
 
```

Es posible sobrescribir los símbolos por defecto del gráfico de dispersión con el argumento markers, tal y como se muestra a continuación. Se puede pasar una selección de símbolos a esta función.

```python
import seaborn as sns

sns.pairplot(df, hue = "species",
             markers = ['d', 'p', 'v']) # los valores d (diamond), p (pentagon) y v (triangle_down) son para cambiar la forma de los puntos
```

## <a id='toc10_4_'></a>[Eliminar el panel superior](#toc0_)

El argumento corner es False por defecto. Si se establece como True se creará un gráfico triangular eliminando el panel superior del gráfico.

```python
import seaborn as sns

sns.pairplot(df, hue = "species",
             corner = True) 
```

Existen varios tipos de gráficos que se pueden crear mediante el argumento **kind**.   
Por defecto, se mostrarán gráficos de dispersión en los paneles con histogramas en la diagonal, pero también puedes seleccionar **"kde"** para estimaciones de densidad tipo núcleo/curvas de nivel, **"hist"** para histogramas y **"reg"** para agregar rectas de regresión sobre las nubes de puntos.

```python
import seaborn as sns

sns.pairplot(df, kind = "kde")
```

## <a id='toc10_5_'></a>[Tipo de diagonal](#toc0_)
La diagonal por defecto de pairplot representa histogramas. Sin embargo, estableciendo el argumento diag_kind como "kde" se puede agregar estimaciones de densidad tipo núcleo/curvas de nivel en lugar de histogramas y estableciendo el argumento como "None" mostrará gráficos de dispersión de cada variable frente a si misma (una línea recta).

```python
import seaborn as sns

sns.pairplot(df, diag_kind = "kde")
```

```python
import seaborn as sns

sns.pairplot(df, diag_kind = None)
```

# <a id='toc11_'></a>[Gráficos marginales ("joinplot")](#toc0_)

Con la función ***jointplot*** es muy sencillo y rápido crear un gráfico de dispersión (y otros tipos de gráficos) con histogramas marginales.    
Tan sólo es necesario pasar los datos a la función para crear la visualización.

```python
import seaborn as sns

df = sns.load_dataset("penguins")

sns.jointplot(data = df,
              x = "bill_length_mm",
              y = "bill_depth_mm")
```

Una alternativa a la función **jointplot** es **JointGrid**, que proporciona mayor flexibilidad. En caso necesario, **JointGrid** es un wrapper de esa función.

```python
import seaborn as sns

df = sns.load_dataset("penguins")

ax = sns.JointGrid(data = df,
                   x = "bill_length_mm",
                   y = "bill_depth_mm")
                   
ax.plot(sns.scatterplot, sns.histplot) 
```

## <a id='toc11_1_'></a>[Recta de regresión y densidades](#toc0_)

La función proporciona un argumento llamado *kind* que permite seleccionar entre distintos tipos de gráficos.    
Las posibles opciones son "scatter" (por defecto), "kde", "hist", "hex", "reg" y "resid".

```python
import seaborn as sns

df = sns.load_dataset("penguins")

sns.jointplot(data = df,
              x = "bill_length_mm", # longitud del pico
              y = "bill_depth_mm",  # grosor del pico
              kind = "reg") #Agrega recta de regresión y densidades 
```

## <a id='toc11_2_'></a>[Tamaño de los gráficos marginales](#toc0_)

Por defecto, el gráfico principal es 5 veces más grande que los gráficos marginales. Se puede cambiar la relación de tamaños con ratio, teniendo en cuenta que este argumento solo acepta números enteros.

```python
import seaborn as sns

df = sns.load_dataset("penguins")

# Gráfico de dispersión dos veces más grande que los histogramas
sns.jointplot(data = df,
              x = "bill_length_mm",
              y = "bill_depth_mm", 
              kind="reg",
              ratio = 2)  
```
