## <a id='toc1_1_'></a>[Colores](#toc0_)

Para cambiar el color de los objetos se utiliza el parámetro **color = nombre-color**, donde **nombre-color** es una cadena con el nombre del color de entre los colores disponibles.

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
dias = ['L', 'M', 'X', 'J', 'V', 'S', 'D']
temperaturas = {'Madrid':[28.5, 30.5, 31, 30, 28, 27.5, 30.5], 'Barcelona':[24.5, 25.5, 26.5, 25, 26.5, 24.5, 25]}
ax.plot(dias, temperaturas['Madrid'], color = 'tab:purple')
ax.plot(dias, temperaturas['Barcelona'], color = 'tab:green')
plt.show()
```

## <a id='toc1_2_'></a>[Marcadores](#toc0_)

Para cambiar la forma de los puntos marcadores se utiliza el parámetro **marker = nombre-marcador** donde **nombre-marcador** es una cadena con el nombre del marcador de entre los marcadores disponibles

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
dias = ['L', 'M', 'X', 'J', 'V', 'S', 'D']
temperaturas = {'Madrid':[28.5, 30.5, 31, 30, 28, 27.5, 30.5], 'Barcelona':[24.5, 25.5, 26.5, 25, 26.5, 24.5, 25]}
ax.plot(dias, temperaturas['Madrid'], marker = '^')
ax.plot(dias, temperaturas['Barcelona'], marker = 'o')
plt.show()
```

## <a id='toc1_3_'></a>[Líneas](#toc0_)

Para cambiar el estilo de las líneas se utiliza el parámetro **linestyle = nombre-estilo** donde **nombre-estilo** es una cadena con el nombre del estilo de entre los estilos disponibles

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
dias = ['L', 'M', 'X', 'J', 'V', 'S', 'D']
temperaturas = {'Madrid':[28.5, 30.5, 31, 30, 28, 27.5, 30.5], 
                'Barcelona':[24.5, 25.5, 26.5, 25, 26.5, 24.5, 25], 
                'Sevilla':[30, 32, 33, 32, 30, 29.5, 32], 
                'Valencia':[26, 27, 28, 27, 26, 25.5, 27],
                'Valladolid':[25, 26, 27, 26, 25, 24.5, 26]}
ax.plot(dias, temperaturas['Madrid'], linestyle = 'dashed')
ax.plot(dias, temperaturas['Barcelona'], linestyle = 'dotted')
ax.plot(dias, temperaturas['Sevilla'], linestyle = 'dashdot')
ax.plot(dias, temperaturas['Valencia'], linestyle = 'solid')
ax.plot(dias, temperaturas['Valladolid'], linestyle = 'dashdot')
plt.show()
```

## <a id='toc1_4_'></a>[Títulos](#toc0_)

Para añadir un título principal al gráfico se utiliza el siguiente método:

**ax.set_title(titulo, loc=alineacion, fontdict=fuente)** : Añade un título con el contenido de la cadena titulo a los ejes *ax*. El parámetro *loc* indica la alineación del título, que puede ser 'left' (izquierda), 'center' (centro) o 'right' (derecha), y el parámetro *fontdict* indica mediante un diccionario las características de la fuente (el tamaño *fontsize*, el grosor *fontweight* o el color *color*).

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
dias = ['L', 'M', 'X', 'J', 'V', 'S', 'D']
temperaturas = {'Madrid':[28.5, 30.5, 31, 30, 28, 27.5, 30.5], 
                'Barcelona':[24.5, 25.5, 26.5, 25, 26.5, 24.5, 25], 
                'Sevilla':[30, 32, 33, 32, 30, 29.5, 32], 
                'Valencia':[26, 27, 28, 27, 26, 25.5, 27],
                'Valladolid':[25, 26, 27, 26, 25, 24.5, 26]}
ax.plot(dias, temperaturas['Madrid'], linestyle = 'dashed')
ax.plot(dias, temperaturas['Barcelona'], linestyle = 'dotted')
ax.plot(dias, temperaturas['Sevilla'], linestyle = 'dashdot')
ax.plot(dias, temperaturas['Valencia'], linestyle = 'solid')
ax.plot(dias, temperaturas['Valladolid'], linestyle = 'dashdot')
ax.set_title('Evolución de la temperatura diaria', loc = "left", fontdict = {'fontsize':14, 'fontweight':'bold', 'color':'tab:blue'})
plt.show()
```

## <a id='toc1_5_'></a>[Ejes](#toc0_)

Para cambiar el aspecto de los ejes se suelen utilizar los siguientes métodos:

- **ax.set_xlabel(titulo)** : Añade un título con el contenido de la cadena titulo al eje x de ax. Se puede personalizar la alineación y la fuente con los mismos parámetros que para el título principal.
- **ax.set_ylabel(titulo)** : Añade un título con el contenido de la cadena titulo al eje y de ax. Se puede personalizar la alineación y la fuente con los mismos parámetros que para el título principal.
- **ax.set_xlim([limite-inferior, limite-superior])** : Establece los límites que se muestran en el eje x de ax.
- **ax.set_ylim([limite-inferior, limite-superior])** : Establece los límites que se muestran en el eje y de ax.
- **ax.set_xticks(marcas)** : Dibuja marcas en el eje x de ax en las posiciones indicadas en la lista marcas.
- **ax.set_yticks(marcas)** : Dibuja marcas en el eje y de ax en las posiciones indicadas en la lista marcas.
- **ax.set_xscale(escala)** : Establece la escala del eje x de ax, donde el parámetro escala puede ser 'linear' (lineal) o 'log' (logarítmica).
- **ax.set_yscale(escala)** : Establece la escala del eje y de ax, donde el parámetro escala puede ser 'linear' (lineal) o 'log' (logarítmica).

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
dias = ['L', 'M', 'X', 'J', 'V', 'S', 'D']
temperaturas = {'Madrid':[28.5, 30.5, 31, 30, 28, 27.5, 30.5], 
                'Barcelona':[24.5, 25.5, 26.5, 25, 26.5, 24.5, 25], 
                'Sevilla':[30, 32, 33, 32, 30, 29.5, 32], 
                'Valencia':[26, 27, 28, 27, 26, 25.5, 27],
                'Valladolid':[25, 26, 27, 26, 25, 24.5, 26]}
ax.plot(dias, temperaturas['Madrid'], linestyle = 'dashed')
ax.plot(dias, temperaturas['Barcelona'], linestyle = 'dotted')
ax.plot(dias, temperaturas['Sevilla'], linestyle = 'dashdot')
ax.plot(dias, temperaturas['Valencia'], linestyle = 'solid')
ax.plot(dias, temperaturas['Valladolid'], linestyle = 'dashdot')
ax.set_title('Evolución de la temperatura diaria', loc = "left", fontdict = {'fontsize':14, 'fontweight':'bold', 'color':'tab:blue'})
ax.set_xlabel("Días", fontdict = {'fontsize':14, 'fontweight':'bold', 'color':'tab:blue'})
ax.set_ylabel("Temperatura ºC")
ax.set_ylim([20,35])
ax.set_yticks(range(20, 35))
plt.show()
```

## <a id='toc1_6_'></a>[Leyenda](#toc0_)

Para añadir una leyenda a un gráfico se utiliza el siguiente método:

- **ax.legend(leyendas, loc = posición)** : Dibuja un leyenda en los ejes ax con los nombres indicados en la lista leyendas.   
 El parámetro *loc* indica la posición en la que se dibuja la leyenda y puede ser 'upper left' (arriba izquierda), 'upper center' (arriba centro), 'upper right' (arriba derecha), 'center left' (centro izquierda), 'center' (centro), 'center right' (centro derecha), 'lower left' (abajo izquierda), 'lower center' (abajo centro), 'lower right' (abajo derecha).    
 Se puede omitir la lista leyendas si se indica la leyenda de cada serie en la función que la dibuja mediante el parámetro label.

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
dias = ['L', 'M', 'X', 'J', 'V', 'S', 'D']
temperaturas = {'Madrid':[28.5, 30.5, 31, 30, 28, 27.5, 30.5], 
                'Barcelona':[24.5, 25.5, 26.5, 25, 26.5, 24.5, 25], 
                'Sevilla':[30, 32, 33, 32, 30, 29.5, 32], 
                'Valencia':[26, 27, 28, 27, 26, 25.5, 27],
                'Valladolid':[25, 26, 27, 26, 25, 24.5, 26]}
ax.plot(dias, temperaturas['Madrid'], linestyle = 'dashed', label = 'Madrid')
ax.plot(dias, temperaturas['Barcelona'], linestyle = 'dotted', label = 'Barcelona')
ax.plot(dias, temperaturas['Sevilla'], linestyle = 'dashdot', label = 'Sevilla')
ax.plot(dias, temperaturas['Valencia'], linestyle = 'solid', label = 'Valencia')
ax.plot(dias, temperaturas['Valladolid'], linestyle = 'dashdot', label = 'Valladolid')
ax.legend(loc = 'upper right') # Leyenda
ax.set_title('Evolución de la temperatura diaria', loc = "left", fontdict = {'fontsize':14, 'fontweight':'bold', 'color':'tab:blue'})
ax.set_xlabel("Días", fontdict = {'fontsize':14, 'fontweight':'bold', 'color':'tab:blue'})
ax.set_ylabel("Temperatura ºC")
ax.set_ylim([20,35])
ax.set_yticks(range(20, 35))
plt.show()
```

## <a id='toc1_7_'></a>[Rejilla](#toc0_)

**ax.grid(axis=ejes, color=color, linestyle=estilo)** : Dibuja una rejilla en los ejes de ax.    
El parámetro **axis** indica los ejes sobre los que se dibuja la regilla y puede ser 'x' (eje x), 'y' (eje y) o 'both' (ambos).    
Los parámetros *color* y *linestyle* establecen el color y el estilo de las líneas de la rejilla, y pueden tomar los mismos valores vistos en los apartados de colores y líneas.

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
dias = ['L', 'M', 'X', 'J', 'V', 'S', 'D']
temperaturas = {'Madrid':[28.5, 30.5, 31, 30, 28, 27.5, 30.5], 
                'Barcelona':[24.5, 25.5, 26.5, 25, 26.5, 24.5, 25], 
                'Sevilla':[30, 32, 33, 32, 30, 29.5, 32], 
                'Valencia':[26, 27, 28, 27, 26, 25.5, 27],
                'Valladolid':[25, 26, 27, 26, 25, 24.5, 26]}
ax.plot(dias, temperaturas['Madrid'], linestyle = 'dashed', label = 'Madrid')
ax.plot(dias, temperaturas['Barcelona'], linestyle = 'dotted', label = 'Barcelona')
ax.plot(dias, temperaturas['Sevilla'], linestyle = 'dashdot', label = 'Sevilla')
ax.plot(dias, temperaturas['Valencia'], linestyle = 'solid', label = 'Valencia')
ax.plot(dias, temperaturas['Valladolid'], linestyle = 'dashdot', label = 'Valladolid')
ax.legend(loc = 'upper right') # Leyenda
ax.set_title('Evolución de la temperatura diaria', loc = "left", fontdict = {'fontsize':14, 'fontweight':'bold', 'color':'tab:blue'})
ax.set_xlabel("Días", fontdict = {'fontsize':14, 'fontweight':'bold', 'color':'tab:blue'})
ax.set_ylabel("Temperatura ºC")
ax.set_ylim([20,35])
ax.set_yticks(range(20, 35))
ax.grid(axis = 'both', color = 'gray', linestyle = 'dotted') # Cuadrícula
plt.show()
```

## <a id='toc1_8_'></a>[Múltiples gráficos](#toc0_)

Es posible dibujar varios gráficos en distintos ejes en una misma figura organizados en forma de tabla. Para ello, cuando se inicializa la figura y los ejes, hay que pasarle a la función **subplots** el número de filas y columnas de la tabla que contendrá los gráficos.    
Con esto los distintos ejes se organizan en un array y se puede acceder a cada uno de ellos a través de sus índices. Si se quiere que los distintos ejes compartan los mismos límites para los ejes se pueden pasar los parámetros **sharex = True** para el eje x o **sharey = True** para el eje y.

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(2, 5, sharey = True, figsize = (20, 10)) # figsize = (ancho, alto) --> Dimensiones de la figura que contiene los gráficos
dias = ['L', 'M', 'X', 'J', 'V', 'S', 'D']
temperaturas = {'Madrid':[28.5, 30.5, 31, 30, 28, 27.5, 30.5], 'Barcelona':[24.5, 25.5, 26.5, 25, 26.5, 24.5, 25], 'Sevilla':[30, 32, 33, 32, 30, 29.5, 32], 'Valencia':[26, 27, 28, 27, 26, 25.5, 27], 'Valladolid':[25, 26, 27, 26, 25, 24.5, 26]}
ax[0, 0].plot(dias, temperaturas['Madrid'])
ax[0, 1].plot(dias, temperaturas['Barcelona'], color = 'tab:orange')
ax[0, 2].plot(dias, temperaturas['Sevilla'], color = 'tab:green')
ax[0, 3].plot(dias, temperaturas['Valencia'], color = 'tab:red')
ax[0, 4].plot(dias, temperaturas['Valladolid'], color = 'tab:purple')

ax[1, 0].bar(dias, temperaturas['Madrid'])
ax[1, 1].bar(dias, temperaturas['Barcelona'], color = 'tab:orange')
ax[1, 2].bar(dias, temperaturas['Sevilla'], color = 'tab:green')
ax[1, 3].bar(dias, temperaturas['Valencia'], color = 'tab:red')
ax[1, 4].bar(dias, temperaturas['Valladolid'], color = 'tab:purple')
plt.show()
```

## <a id='toc1_9_'></a>[INTEGRACIÓN CON PANDAS](#toc0_)

Matplotlib se integra a la perfección con la librería Pandas, permitiendo dibujar gráficos a partir de los datos de las series y DataFrames de Pandas.   


**df.plot(kind=tipo, x=columnax, y=columnay, ax=ejes)** : Dibuja un diagrama del tipo indicado por el parámetro **kind** en los ejes indicados en el parámetro **ax**, representando en el eje x la columna del parámetro x y en el eje y la columna del parámetro y.     
El parámetro **kind** puede tomar como argumentos 'line' (lineas), 'scatter' (puntos), 'bar' (barras verticales), 'barh' (barras horizontales), 'hist' (histograma), 'box' (cajas), 'density' (densidad), 'area' (area) o 'pie' (sectores).     
Es posible pasar otros parámetros para indicar el color, el marcador o el estilo de línea como se vió en los apartados anteriores.

```python
import pandas as pd 
import matplotlib.pyplot as plt
df = pd.DataFrame({'Días':['L', 'M', 'X', 'J', 'V', 'S', 'D'], 
                   'Madrid':[28.5, 30.5, 31, 30, 28, 27.5, 30.5], 
                   'Barcelona':[24.5, 25.5, 26.5, 25, 26.5, 24.5, 25],
                   'Sevilla':[30, 32, 33, 32, 30, 29.5, 32],
                   'Valencia':[26, 27, 28, 27, 26, 25.5, 27],
                   'Valladolid':[25, 26, 27, 26, 25, 24.5, 26]})
fig, ax = plt.subplots()
df.plot(x = 'Días', y = 'Madrid', ax = ax, title='Evolución de la temperatura diaria')
df.plot(x = 'Días', y = 'Barcelona', ax = ax)
df.plot(x = 'Días', y = 'Sevilla', ax = ax)
df.plot(x = 'Días', y = 'Valencia', ax = ax)
df.plot(x = 'Días', y = 'Valladolid', ax = ax)
plt.show()
```

Si no se indican los parámetros **x** e **y**, se representa el índice de las filas en el eje x y una serie por cada columna del Dataframe. Las columnas no numéricas se ignoran.

```python
import pandas as pd 
import matplotlib.pyplot as plt
df = pd.DataFrame({'Días':['L', 'M', 'X', 'J', 'V', 'S', 'D'], 
                   'Madrid':[28.5, 30.5, 31, 30, 28, 27.5, 30.5], 
                   'Barcelona':[24.5, 25.5, 26.5, 25, 26.5, 24.5, 25],
                   'Sevilla':[30, 32, 33, 32, 30, 29.5, 32],
                   'Valencia':[26, 27, 28, 27, 26, 25.5, 27],
                   'Valladolid':[25, 26, 27, 26, 25, 24.5, 26]})
df
```
