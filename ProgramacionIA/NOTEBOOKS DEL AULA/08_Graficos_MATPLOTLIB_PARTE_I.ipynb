# <a id='toc1_'></a>[Creación de gráficos con Matplotlib - Parte I](#toc0_)

Para crear un gráfico con **matplotlib** es común seguir los siguientes pasos:

- Importar el módulo **pyplot**.
- Definir la figura que contendrá el gráfico, que es la region (ventana o página) donde se dibujará y los ejes sobre los que se dibujarán los datos. Para ello se utiliza la función *subplots()*.
- Dibujar los datos sobre los ejes. Para ello se utilizan distintas funciones dependiendo del tipo de gráfico que se quiera.
- Personalizar el gráfico. Para ello existen multitud de funciones que permiten añadir un título, una leyenda, una rejilla, cambiar colores o personalizar los ejes.
- Guardar el gráfico. Para ello se utiliza la función *savefig()*.
- Mostrar el gráfico. Para ello se utiliza la función *show()*.

## <a id='toc1_3_'></a>[Diagramas de dispersión o puntos](#toc0_)

**scatter(x, y)**: Dibuja un diagrama de puntos con las coordenadas de la lista x en el eje X y las coordenadas de la lista y en el eje Y. 

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.scatter([1, 2, 3, 4], [1, 2, 0, 0.5])
plt.show()
```

## <a id='toc1_4_'></a>[Diagramas de líneas](#toc0_)

**plot(x, y)**: Dibuja un polígono con los vértices dados por las coordenadas de la lista x en el eje X y las coordenadas de la lista y en el eje Y. 

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [1, 2, 0, 0.5])
plt.show()
```

## <a id='toc1_5_'></a>[Diagramas de áreas](#toc0_)

**fill_between(x, y)**: Dibuja el area bajo el polígono con los vértices dados por las coordenadas de la lista x en el eje X y las coordenadas de la lista y en el eje Y. 

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.fill_between([1, 2, 3, 4], [1, 2, 0, 0.5])
plt.show()
```

## <a id='toc1_6_'></a>[Diagramas de barras verticales](#toc0_)

**bar(x, y)**: Dibuja un diagrama de barras verticales donde x es una lista con la posición de las barras en el eje X, e y es una lista con la altura de las barras en el eje Y. 

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.bar([1, 2, 3], [3, 2, 1])
plt.show()
```

**hist(x, bins)**: Dibuja un histograma con las frecuencias resultantes de agrupar los datos de la lista x en las clases definidas por la lista **bins**.

## <a id='toc1_12_'></a>[Diagramas de contorno](#toc0_)

**contourf(x, y, z)**: Dibuja un diagrama de contorno con las curvas de nivel de la superficie dada por los puntos con las coordenadas de las listas **x**, **y** y **z** en los ejes X, Y y Z respectivamente. 

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots()
x = np.linspace(-3.0, 3.0, 100)
y = np.linspace(-3.0, 3.0, 100)
x, y = np.meshgrid(x, y)
z = np.sqrt(x**2 + 2*y**2)
ax.contourf(x, y, z)
plt.show()
```

## <a id='toc1_13_'></a>[Mapas de color](#toc0_)

**imshow(x)**: Dibuja un mapa de color a partir de una matriz (array bidimensiona) **x**. 


```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
x = np.random.random((16, 16))
ax.imshow(x)
plt.show()
```

**hist2d(x, y)**: Dibuja un mapa de color que simula un histograma bidimensional, donde los colores de los cuadrados dependen de las frecuencias de las clases de la muestra dada por las listas **x** e **y**. 

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
x, y = np.random.multivariate_normal(mean=[0.0, 0.0], cov=[[1.0, 0.4], [0.4, 0.5]], size=1000).T
ax.hist2d(x, y)
plt.show()
```

## <a id='toc1_14_'></a>[Diagrama de Áreas Apiladas (stackplot())](#toc0_)

Se representa la evolución acumulada de tres categorías a lo largo del tiempo.

**plt.stackplot(x, y1, y2, ..., labels=['Etiqueta1', 'Etiqueta2', ...])**


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 6)
y1 = np.array([1, 3, 4, 8, 10])
y2 = np.array([2, 2, 3, 3, 4])
y3 = np.array([1, 2, 2, 3, 5])

plt.stackplot(x, y1, y2, y3, labels=['Categoría A', 'Categoría B', 'Categoría C'])
plt.legend()
plt.title('Evolución de categorías en el tiempo')
plt.xlabel('Tiempo')
plt.ylabel('Valor acumulado')
plt.show()
```

## <a id='toc1_15_'></a>[Diagrama con Barras de Error (errorbar())](#toc0_)
Se representan mediciones con incertidumbre en los valores.    
Se muestran datos con barras de error indicando la incertidumbre en cada medición.    

**plt.errorbar(x, y, yerr=error, fmt='o', capsize=5)**    

El parámetro *fmt* nos indica el carácter a mostrar como punto central de cada barra


```python
x = np.arange(5)
y = np.array([10, 15, 13, 17, 16])
error = np.array([1, 2, 1, 3, 2])

plt.errorbar(x, y, yerr=error, fmt='o', capsize=5)
plt.title('Gráfico con barras de error')
plt.xlabel('Muestras')
plt.ylabel('Medición')
plt.show()
```

## <a id='toc1_16_'></a>[Mapa de Densidad Hexagonal (hexbin())](#toc0_)

## <a id='toc1_21_'></a>[Gráfico de barras apiladas](#toc0_)

Los gráficos de barras apiladas representan los valores que toman ciertos grupos dentro de otros grupos.    
Por ejemplo, si se considera que personas que viven en cinco ciudades diferentes (esto es, los grupos 'G1', 'G2', 'G3', 'G4' y 'G5') respondieron Si o No a una pregunta y los resultados (el conteo de respuestas en cada ciudad) se guardaron en las variables valores1 y valores2.

Estos datos se pueden representar en Python con la función bar de matplotlib haciendo uso del argumento **bottom**, que representa los valores del eje Y en los que dibujar la parte inferior de las barras, lo que permite apilar unas barras sobre otras.

```python
import matplotlib.pyplot as plt

# Datos
grupos = ['G1', 'G2', 'G3', 'G4', 'G5']
valores1 = [12, 19, 14, 27, 16]
valores2 = [21, 30, 15, 17, 20]

fig, ax = plt.subplots()

# Gráfico de barras apiladas
ax.bar(grupos, valores1)
ax.bar(grupos, valores2, bottom = valores1)

# plt.show() 
 
```

### <a id='toc1_21_1_'></a>[Generalización para varios subgrupos](#toc0_)

Si no se quiere apilar los valores manualmente, se puede unir los valores en un array de numpy y luego utilizar el siguiente bucle for:

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
grupos = ['G1', 'G2', 'G3', 'G4', 'G5']

values = np.array([[12, 19, 14, 27, 16],
                  [21, 30, 15, 17, 20],
                  [15, 23, 12, 11, 15],
                  [2, 5, 1, 6, 8]])

fig, ax = plt.subplots()

# Gráfico de barras apiladas con un bucle
for i in range(values.shape[0]):
  ax.bar(grupos, values[i], bottom = np.sum(values[:i], axis = 0))

# plt.show() 
```

## <a id='toc1_17_'></a>[Gráfico de Vectores (quiver())](#toc0_)

Se visualizan vectores en un espacio bidimensional, con flechas indicando dirección y magnitud.   
Se usa para representar direcciones y magnitudes de fuerzas o velocidades en un espacio bidimensional.

**plt.quiver(X, Y, U, V)**

Parámetros
- X, Y	--> Coordenadas de los puntos donde se dibujarán los vectores. Son matrices generadas por np.meshgrid() para definir la cuadrícula.
- U, V	--> Componentes del vector en las direcciones X e Y. Determinan la dirección y magnitud de cada flecha.
- color	--> Color de las flechas (opcional). Puede ser un solo color o una lista de colores.
- scale	--> Factor de escala para ajustar el tamaño de las flechas. Cuanto mayor sea, más pequeñas se verán las flechas.

```python
# Se crean las posiciones X y Y con np.meshgrid que generan matrices con las posiciones de los puntos
# de dimensiones 5x5, es decir, 25 puntos en total
# Cada np.arange genera un vector con 5 elementos
# meshgrid combina los dos vectores en matrices de 5x5. El formato de meshgrid es el siguiente: 
# X, Y = np.meshgrid([vector en x], [vector en y])
X, Y = np.meshgrid(np.arange(0, 5, 1), np.arange(0, 5, 1)) 
# Se crean los vectores U y V con las componentes de los vectores en cada punto
U = np.cos(X)
V = np.sin(Y)
# Se grafican los vectores que apuntan en la dirección de U y V
# en las posiciones dadas por X y Y 
plt.quiver(X, Y, U, V)
plt.title('Gráfico de vectores')
plt.show()
```

## <a id='toc1_18_'></a>[Diagrama de superficie 3D (plot_surface())](#toc0_)

Se representa una superficie tridimensional basada en una función matemática.     

**ax.plot_surface(X, Y, Z, cmap='plasma')**   

Parámetros   
- X, Y	--> Coordenadas de la cuadrícula en el plano XY. Se generan con np.meshgrid().
- Z	    --> Altura de la superficie en cada punto (X, Y). Es una matriz del mismo tamaño que X y Y.
- cmap	--> Mapa de colores aplicado a la superficie. Ejemplo: 'viridis', 'plasma', 'coolwarm', etc.
- edgecolor	--> Color de los bordes de la superficie. Puede ser 'black', 'white', o None para ocultarlos.

```python
# Se importa la clase Axes3D de la librería mpl_toolkits.mplot3d de matplotlib
from mpl_toolkits.mplot3d import Axes3D 

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d') # Se crea un subplot 3D

# Se crean los datos para la superficie
X, Y = np.meshgrid(np.linspace(-3, 3, 30), np.linspace(-3, 3, 30)) # Se crean las posiciones X y Y, usando meshgrid con 30 puntos en cada dirección 
                                                                   # usando np.linspace que nos genera un vector con 30 elementos entre -3 y 3
Z = np.sin(np.sqrt(X**2 + Y**2)) # Se crea la matriz Z con los valores de la función seno de la raíz cuadrada de X^2 + Y^2

ax.plot_surface(X, Y, Z, cmap='plasma') # Se grafica la superficie con plot_surface y el valor de cmap 'plasma' para el mapa de colores
ax.set_title('Superficie 3D')
plt.show()
```

## <a id='toc1_19_'></a>[Diagrama de Campo de flujo en 2D](#toc0_)

Se utiliza para representar el flujo de un campo vectorial continuo.

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear una malla de coordenadas
x = np.linspace(-3, 3, 20)
y = np.linspace(-3, 3, 20)
X, Y = np.meshgrid(x, y)

# Definir las componentes del campo vectorial (U: dirección X, V: dirección Y)
U = -Y  # Movimiento en X depende de Y
V = X   # Movimiento en Y depende de X

# Crear el gráfico de líneas de corriente
plt.figure(figsize=(6,6))
plt.streamplot(X, Y, U, V, color='b', linewidth=1, density=1)

# Configuración del gráfico
plt.title("Gráfico de Líneas de Corriente con streamplot()")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.grid()
plt.show()
```

### <a id='toc1_19_1_'></a>[Explicación del código:](#toc0_)

1. Definimos una malla de puntos (X, Y): Se usa np.meshgrid() para generar coordenadas en un rango de valores (-3 a 3 en este caso).
2. Creamos un campo vectorial (U, V):
    - U = -Y: El flujo en la dirección X depende de Y.
    - V = X: El flujo en la dirección Y depende de X.    
    Esto crea un flujo que rota en sentido antihorario, simulando un vórtice.
3. Dibujamos las líneas de corriente con plt.streamplot():
    - X, Y: Posiciones en la malla.
    - U, V: Dirección del flujo en cada punto.
    - color='b': Las líneas serán azules.
    - linewidth=1: Grosor de las líneas.
    - density=1: Ajusta la cantidad de líneas en la gráfica.
4. Agregamos etiquetas y rejilla para mejorar la visualización.

## <a id='toc1_20_'></a>[Gráfica de coordenadas polares](#toc0_)

Representación de funciones usando coordenadas polares

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir los ángulos (theta) en radianes
theta = np.linspace(0, 2 * np.pi, 100)  # De 0 a 2π (una vuelta completa)

# Definir el radio en función del ángulo
r = 1 + np.sin(4 * theta)  # Función sinusoidal con 4 oscilaciones

# Crear la figura y el eje polar
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})

# Dibujar la gráfica polar
ax.plot(theta, r, color='r', linewidth=2)

# Configuración del gráfico
ax.set_title("Gráfico Polar de una Función Sinusoidal", va='bottom')

# Mostrar la gráfica
plt.show()
```

### <a id='toc1_20_1_'></a>[Explicación del código:](#toc0_)

1. Definimos los ángulos (theta):
    - Se usa np.linspace(0, 2 * np.pi, 100) para generar 100 puntos equidistantes entre 0 y 360° (2π radianes).
2. Definimos el radio (r):
    - r = 1 + np.sin(4 * theta): Se genera una función sinusoidal con 4 oscilaciones alrededor de un radio base de 1.
3. Creamos el gráfico polar con plt.subplots(subplot_kw={'projection': 'polar'}):
    - Se define un eje de tipo polar para dibujar la gráfica en coordenadas angulares.
4. Trazamos la curva en coordenadas polares con ax.plot(theta, r, color='r', linewidth=2).
5. Agregamos un título y configuramos el gráfico para que sea más legible.

### <a id='toc1_20_2_'></a>[TABLA RESUMEN GRÁFICOS MATPLOTLIB MÁS COMUNES](#toc0_)


| **Tipo de Gráfica**          | **Uso Habitual** | **Ejemplo de Aplicación** |
|-----------------------------|-----------------|--------------------------|
| `plot()` (Gráfica de líneas) | Representación de tendencias en series temporales, funciones matemáticas y evolución de datos. | Seguimiento del precio de una acción en la bolsa a lo largo del tiempo. |
| `bar()` (Gráfico de barras)  | Comparación de cantidades entre diferentes categorías. | Comparación de ventas mensuales de diferentes productos en una tienda. |
| `barh()` (Gráfico de barras horizontales) | Similar al gráfico de barras, pero más útil cuando las etiquetas son largas. | Comparación del PIB per cápita de diferentes países con nombres largos. |
| `hist()` (Histograma)       | Distribución de frecuencias de un conjunto de datos. Útil para analizar distribuciones de datos continuos. | Análisis de la distribución de edades de clientes de un banco. |
| `scatter()` (Diagrama de dispersión) | Relación entre dos variables, detección de correlaciones y patrones. | Relación entre el nivel de educación y el salario promedio de empleados. |
| `pie()` (Gráfico de pastel) | Proporciones relativas de diferentes categorías en un conjunto de datos. | Distribución porcentual del mercado entre varias marcas de teléfonos móviles. |
| `boxplot()` (Diagrama de caja y bigotes) | Análisis estadístico de la distribución de datos, detección de valores atípicos y resumen de datos. | Análisis de la dispersión salarial en diferentes departamentos de una empresa. |
| `stackplot()` (Gráfico de áreas apiladas) | Evolución acumulativa de múltiples variables a lo largo del tiempo. | Distribución del tráfico web en un sitio web entre diferentes fuentes (orgánico, redes sociales, publicidad). |
| `fill_between()` (Gráfico de área) | Visualización de intervalos de confianza o diferencias entre curvas. | Representación de la variabilidad en las temperaturas máximas y mínimas a lo largo del año. |
| `errorbar()` (Gráfico con barras de error) | Representación de datos con sus respectivos márgenes de error. | Resultados de experimentos científicos con mediciones sujetas a incertidumbre. |
| `hexbin()` (Mapa de densidad hexagonal) | Visualización de densidad de puntos en un espacio bidimensional. Similar al diagrama de dispersión pero para grandes volúmenes de datos. | Densidad de ubicaciones GPS de taxis en una ciudad. |
| `imshow()` (Mapa de calor) | Representación de datos matriciales o imágenes en escala de colores. | Matriz de correlación entre múltiples variables en un análisis estadístico. |
| `contour()` (Gráfico de contornos) | Representación de curvas de nivel en datos tridimensionales o superficies. | Representación de curvas de nivel en un mapa topográfico. |
| `quiver()` (Gráfico de vectores) | Visualización de campos vectoriales, como dirección y magnitud del viento o fuerzas en un espacio. | Representación de la velocidad y dirección del viento en un mapa meteorológico. |
| `streamplot()` (Gráfico de líneas de corriente) | Representación del flujo de un campo vectorial continuo. | Simulación de flujo de aire alrededor de un ala de avión. |
| `polar()` (Gráfico polar) | Datos que dependen de ángulos, como patrones de radiación o trayectorias circulares. | Representación de la intensidad de la señal de una antena en diferentes direcciones. |
| `3D plot` (`plot_surface()`, `scatter3D()`, etc.) | Visualización de datos en tres dimensiones para modelos complejos o análisis espaciales. | Modelo tridimensional del relieve de una región geográfica. |
